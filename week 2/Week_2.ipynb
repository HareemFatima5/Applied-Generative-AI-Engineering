```python
import google.generativeai as genai
import getpass

API_key = getpass.getpass("Enter your api here:")
genai.configure(api_key=API_key)
model=genai.GenerativeModel("gemini-3.1-flash-lite-preview")
```

### **Zero-Shot Prompting**

We want AI to answer without giving it any examples

### **Example 1**

```python
prompt="""Translate this sentence to Urdu
Sentence: I am learning AI"""

response = model.generate_content(prompt)
print("Zero shot prompting")
print(response.text)
```

### **Example 2**

```python
prompt="""Classify this review as Positive or Negative:
'The product was amazing and delivery was fast."""

response = model.generate_content(prompt)
print("Zero shot prompting")
print(response.text)
```

### **Example 3**

```python
prompt="""Summarize this paragraph
Reading books is an incredibly rewarding habit that broadens our horizons and sharpens our minds."""

response = model.generate_content(prompt)
print("Zero shot prompting")
print(response.text)
```

### **Few-Shot Prompting**

We give model some examples so it unerstand what user wants

### **Example 1**

```python
prompt = """Here are some examples of sentiment classification:

Review: "The food was delicious and the staff was so friendly!" = Positive
Review: "Waited 2 hours and the food was cold. Never again." = Negative
Review: "It was fine, nothing too special about it."  Neutral

Now classify this:
Review: "Honestly the best coffee I have ever had in my life!" ="""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Example 2**

```python
prompt = """Look at these emails and their categories:

Email: "URGENT: Server is down, please fix immediately!" = Urgent
Email: "Hey, are we still on for lunch tomorrow?" = Normal
Email: "Congratulations! You won a free iPhone. Click here." = Spam

Now classify this:
Email: "Please review and approve the report before 5pm today." ="""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Example 3**

```python
prompt = """Here are some job titles and their departments:

Job Title: "Surgeon" = Department: Healthcare
Job Title: "Software Engineer" = Department: Technology
Job Title: "Chartered Accountant" = Department: Finance

Now classify this:
Job Title: "Data Scientist" ="""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Role-Based Prompting**

We assign the model a role so it acts accordingly

### **Example 1**

```python
prompt = """You are a Python teacher who has been teaching beginners for 10 years.One of your students does not understand what a loop is.
Explain what a for loop is in Python using a very simple real life example. Then show a basic code example."""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Example 2**

```python
prompt = """You are an experienced doctor who explains medical conditions in very simple language for patients who have no medical background.
Explain what diabetes is, what causes it, and what a patient should avoid eating. Keep it simple and friendly."""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Example 3**

```python
prompt = """You are a senior hiring manager at a top tech company.
You are interviewing a fresh computer science graduate for a junior data science position.
Ask them 5 interview questions, one by one, starting from easy and getting harder."""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Output constraints**

We add some rules so model answers accordingly

### **Example 1**

```python
prompt = """List 5 benefits of drinking water daily.
Rules you must follow:
1) Use bullet points only
2) Each point must be under 10 words
3) Do not number the points"""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Example 2**

```python
prompt = """Explain artificial intelligence in exactly 50 words not more, not less.
Use simple language that a 10 year old can understand.Do not use any technical terms."""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Example 3**

```python
prompt = """Give me information about Python programming language.
You must follow this exact format and no other:
Name:
Created In:
Created By:
Main Use:
Do not add anything outside this format."""

response = model.generate_content(prompt)
print("Output:", response.text)
```
