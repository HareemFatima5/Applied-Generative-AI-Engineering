```python
import google.generativeai as genai
import getpass

API_key = getpass.getpass("Enter your api here:")
genai.configure(api_key=API_key)
model=genai.GenerativeModel("gemini-3.1-flash-lite-preview")
```

### **Zero-Shot Prompting**

We want AI to answer without giving it any examples

### **Example 1**

```python
prompt="""Translate this sentence to Urdu
Sentence: I am learning AI"""

response = model.generate_content(prompt)
print("Zero shot prompting")
print(response.text)
```

### **Example 2**

```python
prompt="""Classify this review as Positive or Negative:
'The product was amazing and delivery was fast."""

response = model.generate_content(prompt)
print("Zero shot prompting")
print(response.text)
```

### **Example 3**

```python
prompt="""Summarize this paragraph
Reading books is an incredibly rewarding habit that broadens our horizons and sharpens our minds."""

response = model.generate_content(prompt)
print("Zero shot prompting")
print(response.text)
```

### **Few-Shot Prompting**

We give model some examples so it unerstand what user wants

### **Example 1**

```python
prompt = """Here are some examples of sentiment classification:

Review: "The food was delicious and the staff was so friendly!" = Positive
Review: "Waited 2 hours and the food was cold. Never again." = Negative
Review: "It was fine, nothing too special about it." = Neutral

Now classify this:
Review: "Honestly the best coffee I have ever had in my life!" ="""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Example 2**

```python
prompt = """Look at these emails and their categories:

Email: "URGENT: Server is down, please fix immediately!" = Urgent
Email: "Hey, are we still on for lunch tomorrow?" = Normal
Email: "Congratulations! You won a free iPhone. Click here." = Spam

Now classify this:
Email: "Please review and approve the report before 5pm today." ="""

response = model.generate_content(prompt)
print("Output:", response.text)
```

### **Hallucination Control**
we ask the model to ground its answer in knowledge provided and not make up things

### **Example 1**

```python
text = "The Eiffel Tower is located in Paris, France. It was built in 1889 and stands 330 meters tall. It was designed by Gustave Eiffel."

prompt = ("Answer the question below using ONLY the text provided.\n"
        "If the answer is not in the text, say I don't have enough information.\n"
         "Do NOT use outside knowledge.\n\n"
         "Text: " + text +
         "\n\nQuestion: When was the Eiffel Tower built and who designed it?")

response = model.generate_content(prompt)
print("Output:")
print(response.text)
```

### **Example 2**

```python
text = "Hareem Fatima is a data scientist at TechSoft Solutions. She has 3 years of experience working with Python and machine learning models."

prompt = ("Answer the question below using ONLY the text provided.\n"
    "If the answer is not in the text, say I don't have enough information.\n"
    "Do NOT guess or add anything extra.\n\n"
    "Text: " + text +
    "\n\nQuestion: What university did Hareem Fatima attend?")

response = model.generate_content(prompt)

print("Output:")
print(response.text)
```

### **Example 3**

```python
text = "The 2024 Paris Olympics opened on July 26th. Over 10,000 athletes from 206 countries participated. The opening ceremony was held along the Seine River."

prompt = ("Answer the question below using ONLY the text provided.\n"
    "After your answer, write Confidence: High, Medium or Low depending on how sure you are.\n"
    "If the answer is not in the text, say I don't have enough information.\n\n"
    "Text: " + text +
    "\n\nQuestion: How many countries participated in the 2024 Paris Olympics?")

response = model.generate_content(prompt)

print("Output:")
print(response.text)
```
